## Simple Linear Regression task
This is a regression task, we will predict the percentage a student is expected to score based upon the number of hours they have studied. This involves just two variables so this is a simple linear regression task .

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt  
```

### Importing the data

```python
url = "http://bit.ly/w-data"
df = pd.read_csv(url)
print(df.shape)
df.head()
```

### Let's plot the graph

```python
sns.scatterplot(data=df, x="Hours", y="Scores")
plt.title('Hours vs Percentage') 
plt.grid()
plt.legend(["blue"], loc ="upper left")
plt.show()
```

##### As we can see, percentage is linearly increasing with the number of hours they have studied.

Now lets split the data set into train and test, We'll do this by using Scikit-Learn's built-in train_test_split() method and we will use LinearRegression algorithm from Scikit-learn to train the model.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['Hours']], df['Scores'], test_size=0.2, random_state=0) 
```

### Training the algorithm

```python
model = LinearRegression()
model.fit(X_train, y_train)
print("Training complete")
```

### Making Predictions
Now it's time to make some predictions.

```python
y_predict = model.predict(X_test)
```

```python
# We will check what will be the prediction if they studied for 9.5hrs.

given_hour = 9.5
pred = model.predict([[given_hour]])
print("No of Hours = {} and the predicted score for it will be = {} ".format(given_hour, pred[0]))
```

### **Evaluating the model**

Let's evaluate our algorithm's performance. For simplicity here, we have chosen the mean square error.

```python
from sklearn import metrics  
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_predict))
```
